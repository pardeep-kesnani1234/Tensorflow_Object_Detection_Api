```python
import cv2

# Opens the Video file
cap= cv2.VideoCapture('jerry.mp4')
i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imwrite('images/'+str(i)+'.jpg',frame)
    i+=1

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os




pathIn = 'jerry.mp4'
pathOut = "images"
name = ""
int_desiredLength=3
sec = 1.5


if not os.path.exists(pathOut):
    os.makedirs(pathOut)


if os.path.exists(pathIn):

    cap = cv2.VideoCapture(pathIn)
    fps = cap.get(cv2.CAP_PROP_FPS)
    fps = fps
    fps_2_capture = round(sec * fps)

    actual_frame = 0
    vid_frame_saved = 0

    while (cap.isOpened()):

        ret, frame = cap.read()

        #get current frame
        frameId = round(cap.get(1))



        if not ret == True:
            break

        if frameId % fps_2_capture == 0:

            vid_frame = 'frame' + str(actual_frame)
            #str_0 conv int to str
            fname = name + '_{str_0:0>{str_1}}.jpg'.format(str_0=vid_frame_saved, str_1=int_desiredLength)

            print('Extracting {} to: {}'.format(vid_frame , fname))
            cv2.imwrite(os.path.join(pathOut, fname), frame)

            vid_frame_saved += 1

    actual_frame += 1

else:
    print('path to video file does not exist')

print('Finished Extracting')

cap.release()
cv2.destroyAllWindows()
```

```python
import os
import glob
import cv2

fnames = glob.glob(os.path.join("object detection api/images", "*.{}".format("jpg")))
```

```python
for i, fname in enumerate(fnames):
    print(".", end="", flush=True)
    img = cv2.imread(fname)	
    #print(fname)
    img_small = cv2.resize(img, (300,300))
    new_fname = fname.split("\\")[1]
    small_fname = os.path.join("object detection api/resized", new_fname)
    cv2.imwrite(small_fname, img_small)
    print("Done resizing {} files.\nSaved to directory: `{}`".format(len(fnames), "object detection api/resized"))
```
